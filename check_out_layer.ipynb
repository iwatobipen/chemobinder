```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs
from rdkit import RDConfig
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
import numpy as np
plt.style.use('ggplot')
```

```python
sdf = os.path.join(RDConfig.RDDocsDir, 'Book/data/cdk2.sdf')
```

```python
mols = [m for m in Chem.SDMolSupplier(sdf)]
for mol in mols:
    AllChem.Compute2DCoords(mol)
```

```python
# Check prop names
print(list(mols[0].GetPropsAsDict().keys()))
print(len(mols))
print(len(set([mol.GetProp('Cluster') for mol in mols])))
Draw.MolsToGridImage(mols[:5], molsPerRow=5)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.decomposition import PCA
```

```python
fps = [list(AllChem.GetMorganFingerprintAsBitVect(mol, 2)) for mol in mols]
```

```python
fps = np.array(fps)
```

```python
pca = PCA(n_components=2)
pcares = pca.fit_transform(fps)
plt.scatter(pcares[:,0], pcares[:,1], c='b', alpha=0.4)
```

```python
isoforest = IsolationForest(contamination='legacy')
```

```python
isoforest_res = isoforest.fit_predict(pcares)
```

```python
print(isoforest_res)
```

```python
xx, yy = np.meshgrid(np.linspace(-5,5,50), np.linspace(-5,5,50))
```

```python
cmap = {1:'blue', -1:'red'}
cs = [cmap[k] for k in isoforest_res]
z = isoforest.decision_function(np.c_[xx.ravel(), yy.ravel()])
print(z.shape)
print(xx.shape)
z = z.reshape(xx.shape)
```

```python

plt.title("IsolationForest")
plt.contourf(xx, yy, z, cmap=plt.cm.Blues_r)

b1 = plt.scatter(pcares[:, 0], pcares[:, 1], s=20, edgecolor='k', c=cs)
plt.axis('tight')
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.show()
```

```python
outliermols = []
inmols = []
for i, mol in enumerate(mols):
    if isoforest_res[i] == -1:
        outliermols.append(mol)
    else:
        inmols.append(mol)
```

```python
Draw.MolsToGridImage(outliermols, molsPerRow=5)
```

```python
Draw.MolsToGridImage(inmols, molsPerRow=5)
```

```python
# from sklearn example from URL below
# https://scikit-learn.org/stable/auto_examples/ensemble/plot_isolation_forest.html
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

rng = np.random.RandomState(42)

# Generate train data
X = 0.3 * rng.randn(100, 2)
X_train = np.r_[X + 2, X - 2]
# Generate some regular novel observations
X = 0.3 * rng.randn(20, 2)
X_test = np.r_[X + 2, X - 2]
# Generate some abnormal novel observations
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))

# fit the model
clf = IsolationForest(behaviour='new', max_samples=100,
                      random_state=rng, contamination='auto')
clf.fit(X_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
y_pred_outliers = clf.predict(X_outliers)

# plot the line, the samples, and the nearest vectors to the plane
xx, yy = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.title("IsolationForest")
plt.contourf(xx, yy, Z, cmap=plt.cm.Blues_r)

b1 = plt.scatter(X_train[:, 0], X_train[:, 1], c='white',
                 s=20, edgecolor='k')
b2 = plt.scatter(X_test[:, 0], X_test[:, 1], c='green',
                 s=20, edgecolor='k')
c = plt.scatter(X_outliers[:, 0], X_outliers[:, 1], c='red',
                s=20, edgecolor='k')
plt.axis('tight')
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.legend([b1, b2, c],
           ["training observations",
            "new regular observations", "new abnormal observations"],
           loc="upper left")
```
